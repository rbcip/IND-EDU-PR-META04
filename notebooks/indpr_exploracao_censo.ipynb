```python
import pandas as pd
import os
from sqlalchemy import create_engine
import sys
from sqlalchemy.engine import URL
from sqlalchemy import text
```

```python
USER = os.getenv('DB_USER_IND', 'postgres')
PASSWORD = os.environ.get('DB_PASSWORD_IND', '')
HOST = os.environ.get('DB_HOST_IND', 'localhost')
```

```python
url_object = URL.create(
    "postgresql+psycopg2",
    username=USER,
    password=PASSWORD,  # plain (unescaped) text
    host=HOST,
    database="indicadores",
)

engine = create_engine(url_object)
```

```python
dados = pd.read_csv('../dados/censoescolar/microdados_ed_basica_2007.csv', encoding='ISO-8859-1', sep=';', engine='c', low_memory=False)
dados
```

```python
dados2 = pd.read_csv('../dados/censoescolar/microdados_ed_basica_2023.csv', encoding='ISO-8859-1', sep=';', engine='c', low_memory=False)
dados2
```

```python
dados.shape[0] + dados2.shape[0]
```

```python
def agrupa_arquivos(df1, df2):
    for col in df1.columns:
        if col not in df2.columns:
            df2[col] = None
            
    for col in df2.columns:
        if col not in df1.columns:
            df1[col] = None

    return pd.concat([df1, df2])
    
```

```python
df = agrupa_arquivos(dados, dados2)
```

```python
len(set(list(dados.columns) + list(dados2.columns)))
```

```python
df.shape
```

```python
df.to_sql("microdados_ed_basica", schema='stage', con=engine)
```

```python
df = pd.read_csv('../dados/csv/censoescolar/microdados_ed_basica_2007.csv', encoding='ISO-8859-1', sep=';', engine='pyarrow')
```

```python
descricao = 'censoescolar'
table = 'microdados_ed_basica_2007'
schema = 'stage'
filename = 'microdados_ed_basica_2007.csv'
sep = ';'
df.iloc[0:1,:].to_sql(table, schema=schema, con=engine, index=False)
```

```python
copy = f"""
    COPY {schema}.{table}
    FROM '/mnt/{descricao}/{filename}'
    DELIMITER '{sep}'
    ENCODING 'ISO-8859-1'
    CSV HEADER;
"""
```

```python
with engine.connect() as connection:
    result = connection.execute(text(f"truncate table {schema}.{table}"))
    connection.commit()
    result = connection.execute(text(copy))
    connection.commit()
```
